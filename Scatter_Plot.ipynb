## Scatter Plot --------------------------

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
day=[1,2,3,4,5,6,7]
no=[3,5,7,3,2,9,6]

plt.title("Scatter Plot", fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Number",fontsize=15)

plt.scatter(day,no)
plt.show()
```

```python
day=[1,2,3,4,5,6,7]
no=[3,5,7,3,2,9,6]

plt.title("Scatter Plot", fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Number",fontsize=15)

size=[69,70,100,174,87,89,90]
# color=['r','y','g','b','r','b','g']
color=[10,23,40,50,56,43,66]

plt.scatter(day,no,s=size,c=color,cmap="OrRd")
p=plt.colorbar()
p.set_label("ColorBar",fontsize=15)
# plt.scatter(day,no,s=size,c=color, alpha=0.5,marker="*", edgecolor="g", linewidth=4)
plt.show()
```
