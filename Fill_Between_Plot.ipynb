## Fill_Between_Plot -----------------------

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,3,4,5,6]

plt.plot(x,y,color="r")

plt.fill_between(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,3,4,5,6])

plt.plot(x,y,color="r")
plt.fill_between(x,y,where=(x>=2) & (x<=4), alpha=0.4)
plt.show()
```
