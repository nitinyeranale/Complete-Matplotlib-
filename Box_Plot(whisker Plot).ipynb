### Box Plot  ( Whisker Plot ) --------------------

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,80,130]
plt.boxplot(x,notch=False,vert=True)
plt.show()
```

```python
x=[10,20,30,40,50,80,130]
plt.boxplot(x,notch=False,widths=0.3)
plt.show()
```

```python
x=[10,20,30,40,50,80,130]
plt.boxplot(x,notch=False,tick_labels=["Python"],patch_artist=False,showmeans=True)
plt.show()
```

```python
x=[10,20,30,40,50,80,130]
plt.boxplot(x,whis=1.7)
plt.show()
```

```python
x=[10,20,30,40,50,80,130]
x1=[15,30,45,60,75,90,105]

y=[x,x1]

plt.boxplot(y,tick_labels=["Python","C"])
plt.show()
```
