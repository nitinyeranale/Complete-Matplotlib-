### SubPlot-----------------

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color="r")

plt.pie([1])

x=[20,40,50,10]
plt.pie(x)

x=["A","B","C","D"]
y=[30,24,54,75]
plt.bar(x,y)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.subplot(2,3,1)
plt.plot(x,y,color="r")

plt.subplot(2,3,2)
plt.pie([1])

plt.subplot(2,3,3)
x=[20,40,50,10]
plt.pie(x)

plt.subplot(2,3,4)
x=["A","B","C","D"]
y=[30,24,54,75]
plt.bar(x,y)

plt.subplot(2,3,5)
day=[1,2,3,4,5,6,7]
no=[4,2,5,6,8,4,9]
plt.scatter(day,no)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.subplot(1,2,1)
plt.plot(x,y,color="r")

plt.subplot(1,2,2)
x=[20,40,50,10]
plt.pie(x)
plt.show()
```
