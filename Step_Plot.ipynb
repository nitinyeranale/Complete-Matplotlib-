### Step Plot ---------------------

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[11,33,55,67,88,98]
plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[11,33,55,67,88,98]

plt.title("Step Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()

plt.step(x,y,marker="o",ms=8,mfc="r",color='k')
plt.show()
```
