### How to Add Text in Plot --------------

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,1,4,2,3]

plt.plot(x,y)

plt.title("Text_Add",fontsize=15)
plt.xlabel("X_Axis",fontsize=15)
plt.ylabel("Y_Axis",fontsize=15)

plt.text(3,2,"This is text",style="italic",fontweight="bold",bbox={"facecolor":"y"})

plt.annotate("Python",xy=(2,1),xytext=(4,1.5),arrowprops=dict(facecolor="r"))

plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,1,4,2,3]

plt.plot(x,y)

plt.title("Text_Add",fontsize=15)
plt.xlabel("X_Axis",fontsize=15)
plt.ylabel("Y_Axis",fontsize=15)

plt.text(3,2,"This is text",style="italic",fontweight="bold",bbox={"facecolor":"y"})

plt.annotate("Python",xy=(2,1),xytext=(4,1.5),arrowprops=dict(facecolor="r"))

plt.legend(["up"],loc=2,facecolor="g",edgecolor="r",shadow=True)

plt.show()
```
