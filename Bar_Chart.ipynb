```python
import matplotlib.pyplot as plt
import numpy as np
```

## Bar Plot ---------------------------------

```python
x=["Python","Java","C","R"]
y=[95,80,79,92]

plt.xlabel("Language",fontsize=20)
plt.ylabel("Rank",fontsize=20)
plt.title("Bar_Chart",fontsize=20)

plt.bar(x,y,width=0.4, label="Popularity")

plt.legend()
plt.show()
```

```python
x=["Python","Java","C","R"]
y=[95,80,79,92]
z=[67,59,69,66]

width=0.2
p=np.arange(len(x))
p1=[i+width for i in p]

plt.xlabel("Language",fontsize=20)
plt.ylabel("Rank",fontsize=20)
plt.title("Bar_Chart",fontsize=20)


plt.bar(p,y,width, label="Popularity")
plt.bar(p1,z,width, label="Popularity1")

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
year=["2017","2019","2021","2023"]
cs=[89,87,67,59]
it=[67,56,86,44]
ec=[56,98,34,67]
w=0.2

cs_bar=np.arange(len(year))
it_bar=[i+w for i in cs_bar]
ec_bar=[i+w for i in it_bar]

plt.xlabel("Y E A R S")
plt.ylabel("Subject")
plt.title("New_Bar")

plt.bar(cs_bar,cs,width=w,label="cs")
plt.bar(it_bar,it,width=w,label="it")
plt.bar(ec_bar,ec,width=w,label="ec")

plt.xticks(cs_bar+w,year)
plt.legend()
plt.show()
```

```python
birth=[1998,1999,2000,2001]
boy=[80,78,67,34]
girl=[67,98,23,56]
w=0.4

boy_bar=np.arange(len(birth))
girl_bar=[i+w for i in boy_bar]

plt.xlabel("Birth Year")
plt.ylabel("Ratio")
plt.title("Boy and Girl ratio with Year")
plt.bar(boy_bar,boy, width=w, label="Boy")
plt.bar(girl_bar,girl, width=w, label="Girl")

plt.xticks(boy_bar+w/2,birth)
plt.legend()
plt.show()
```

```python
x=["Python","Java","C"]
y=[98,77,45]
z=[87,76,88]
w=0.2

y_bar=np.arange(len(x))
z_bar=[i+w for i in y_bar]

plt.xlabel("Is X")
plt.ylabel("Is Y")
plt.title("Demo")

plt.bar(y_bar,y,width=w)
plt.bar(z_bar,z, width=w)

plt.xticks(y_bar+w/2,x)
plt.show()
```
