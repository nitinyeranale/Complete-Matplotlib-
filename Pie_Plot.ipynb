### Pie Plot ------------------------

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50]
y=["c","c++","Java","R","Python"]
ec=[0.3,0,0,0,0]
plt.pie(x,labels=y,explode=ec)
plt.show()
```

```python
x=[10,20,30,40,50]
y=["c","c++","Java","R","Python"]
ec=[0,0,0.2,0,0]
c=["r","g","b","y","c"]
plt.pie(x,labels=y,explode=ec, autopct="%0.2f%%",shadow=True,radius=1.2,labeldistance=0.8,startangle=90)
plt.show()
```

```python
x=[10,20,30,40,50]
y=["c","c++","Java","R","Python"]
ec=[0.3,0,0,0,0]
plt.pie(x,labels=y,explode=ec,autopct="%0.2f%%",textprops={"fontsize":15,"color":"m"},wedgeprops={"linewidth":3,"edgecolor":"b"})
plt.show()
```

```python
x=[10,20,30,40,50]
y=["c","c++","Java","R","Python"]
ec=[0.3,0,0,0,0]
plt.pie(x,labels=y,explode=ec,autopct="%0.2f%%",radius=0.8)
plt.legend(loc="upper right")
plt.show()
```

```python
plt.pie([1])
plt.show()
```

#### Donut Pie chart---------

```python
x=[10,20,30,40,50]
# x1=[40,58,34,22,57]
y=["c","c++","Java","R","Python"]


plt.pie(x,labels=y)
# plt.pie(x,labels=y,radius=0.2)
plt.pie([1],colors="w",radius=0.6)

# plt.legend(loc="upper right")

plt.show()
```

```python
x=[10,20,30,40,50]
y=["c","c++","Java","R","Python"]

plt.pie(x,labels=y,radius=1.5)
c=plt.Circle((0,0),radius=1,facecolor="w")
plt.gca().add_artist(c)

# plt.legend(loc="upper right")

plt.show()
```

```python
x=[98,76,57,87,44]
x_label=["A","B","C","D","E"]

y=[67,87,54,89,76]
y_label=["M","N","O","P","Q"]

plt.pie(x,labels=x_label,radius=1,wedgeprops={"edgecolor":"black"},labeldistance=0.8)
plt.pie(y,labels=y_label,radius=0.7,wedgeprops={"edgecolor":"black"},labeldistance=0.7)

C=plt.Circle((0,0),radius=0.4,facecolor="w")
plt.gca().add_artist(C)


plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for outer ring
sizes_outer = [30, 10, 20, 40]
labels_outer = ['A1', 'B1', 'C1', 'D1']
colors_outer = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

# Data for inner ring
sizes_inner = [15, 15, 5, 5, 10, 10, 20, 20]  
labels_inner = ['A2', 'A3', 'B2', 'B3', 'C2', 'C3', 'D2', 'D3']
colors_inner = ['#e60000','#ff4d4d','#005ce6','#66a3ff',
                '#009900','#66cc66','#ff9900','#ffcc66']

# Plot outer ring
plt.pie(sizes_outer, labels=labels_outer, colors=colors_outer, 
        radius=1, wedgeprops={'edgecolor': 'black'})

# Plot inner ring
plt.pie(sizes_inner, labels=labels_inner, colors=colors_inner, 
        radius=0.7, wedgeprops={'edgecolor': 'black'})

# Add white circle in the center
center_circle = plt.Circle((0, 0), 0.4, fc='white')
plt.gca().add_artist(center_circle)

# Show plot
plt.show()
```
