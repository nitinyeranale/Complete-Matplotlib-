### Stack Plot or Area Plot -------------------------

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
area=[3,2,1,5,7,3]
area1=[3,5,2,3,6,8]
area2=[4,2,6,6,4,2]

plt.stackplot(x,area,area1,area2)
plt.show()
```

```python
x=[1,2,3,4,5,6]
area=[3,2,1,5,7,3]
area1=[3,5,2,3,6,8]
area2=[4,2,6,6,4,2]
L=["A1","A2","A3"]

plt.stackplot(x,area,area1,area2,labels=L,colors=["r","y","g"])
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,6]
area1=[3,2,1,5,7,3]
area2=[3,5,2,3,6,8]
area3=[4,2,6,6,4,2]
# L=["A1","A2","A3"]

plt.stackplot(x,area1,labels="Area 1",colors="r")
plt.stackplot(x,area2,labels="Area 2",colors="y")
plt.stackplot(x,area3,labels="Area 3",colors="g")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 6)
y1 = [3, 5, 2, 6, 4]
y2 = [1, 2, 4, 3, 5]
y3 = [2, 3, 5, 4, 2]

# Create stack plot
plt.stackplot(x, y1, y2, y3, labels=['A', 'B', 'C'], alpha=0.6)

# Overlay line plots
plt.plot(x, y1, color='red', marker='o', linestyle='--', label="A Trend")
plt.plot(x, y2, color='green', marker='s', linestyle='-.', label="B Trend")
plt.plot(x, y3, color='blue', marker='d', linestyle='-', label="C Trend")

# Labels & Title
plt.xlabel("Time")
plt.ylabel("Values")
plt.title("Stack Plot with Line Overlays")
plt.legend()

# Show
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 6)
y1 = [3, 5, 2, 6, 4]
y2 = [1, 2, 4, 3, 5]
y3 = [2, 3, 5, 4, 2]

# Custom colors
colors = ['red', 'green', 'blue']

# Create stack plot
plt.stackplot(x, y1, y2, y3, labels=['Product A', 'Product B', 'Product C'], colors=colors, alpha=0.6)

# Labels & Title
plt.xlabel("Years")
plt.ylabel("Sales")
plt.title("Customized Stack Plot")
plt.legend()

# Show
plt.show()
```
