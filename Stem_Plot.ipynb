### Stem Plot -------------------------

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,1,6,3,5,2]
plt.stem(x,y,linefmt="g--",markerfmt="r*",bottom=0.5,basefmt="b--")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,1,6,3,5,2]
plt.stem(x,y,linefmt="g--",markerfmt="r*",bottom=0.5,basefmt="b--",label="Python")
plt.legend()
plt.show()
```
